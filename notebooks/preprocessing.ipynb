# 🧠 Bangla Sentiment Analysis: Preprocessing
This notebook covers the data cleaning, normalization, and exploration steps required before training our models.

```python
import pandas as pd
import numpy as np
import re
import matplotlib.pyplot as plt
import seaborn as sns
import emoji
import os

# Setting plot style
plt.style.use('ggplot')
sns.set_palette('husl')
```

## 1. Load Dataset

```python
df = pd.read_csv('../data/annotated_data.csv')
print(f"Dataset Shape: {df.shape}")
df.head()
```

## 2. Exploratory Data Analysis (EDA)

```python
# Distribution of Labels
plt.figure(figsize=(8, 6))
df['label'].value_counts().plot(kind='bar', color=['skyblue', 'salmon', 'lightgreen'])
plt.title('Sentiment Distribution')
plt.xlabel('Sentiment')
plt.ylabel('Count')
plt.show()
```

```python
# Platform-wise Sentiment
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='platform', hue='label')
plt.title('Sentiment across Platforms')
plt.show()
```

## 3. Data Cleaning & Normalization

```python
def clean_text(text):
    # Remove URLs
    text = re.sub(r'http\S+|www\S+|https\S+', '', text, flags=re.MULTILINE)
    # Remove Punctuation but keep emojis for now
    text = re.sub(r'[^\w\s\u0980-\u09FF\u2600-\u27BF\U0001f300-\U0001f64f\U0001f680-\U0001f6ff]', '', text)
    # Remove numbers
    text = re.sub(r'\d+', '', text)
    return text.strip()

df['cleaned_text'] = df['original_text'].apply(clean_text)
df[['original_text', 'cleaned_text']].head()
```

## 4. Handling Lexicons (Stopwords & Slangs)

```python
# Load Stopwords
with open('../lexicon/stopwords.txt', 'r', encoding='utf-8') as f:
    stopwords = set([line.strip() for line in f])

def remove_stopwords(text):
    words = text.split()
    filtered_words = [w for w in words if w not in stopwords]
    return " ".join(filtered_words)

df['final_text'] = df['cleaned_text'].apply(remove_stopwords)
df[['cleaned_text', 'final_text']].head()
```

## 5. Save Processed Data

```python
df.to_csv('../data/processed_data.csv', index=False)
print("Processed data saved to data/processed_data.csv")
```
